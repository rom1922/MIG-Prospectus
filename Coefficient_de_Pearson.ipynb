```python
import numpy as np
def pearson_coef(n,X,Y): #n longueur échantillon, X, Y les deux sets de données à comparer
    mean_x=np.mean(np.array(X))
    mean_y=np.mean(np.array(Y))
    s_x,s_y=0,0
    for i in range(n):
        s_x+=(X[i]-mean_x)**2
        s_y+=(Y[i]-mean_y)**2
    s_x=np.sqrt(s_x/(n-1))
    s_y=np.sqrt(s_y/(n-1))
    p=0
    for i in range(n):
        p+=(X[i]-mean_x)/s_x*(Y[i]-mean_y)/s_y
    p=p/(n-1)
    return p
#Autrement avec .corr avec pandas. 
```
